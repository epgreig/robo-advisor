```python
def clean_float(df):
    
    sub_df = df.copy()
    
    for col in df.columns.values:
        try:
            sub_df[col] = sub_df[col].astype(float)
        except ValueError:
            pass
    
    return sub_df
```

```python
df = clean_float(df)
df_grp = clean_float(df_grp)
grp_list = df_grp.drop(columns=['Name', 'Ticker']).columns.values
```

```python
df = df.merge(df_grp, how='inner', on=['Ticker', 'Name'])
```

```python
df.set_index('Ticker', inplace=True)
```

```python
grp_list
```

```python
grp_etfs = []
for grp in grp_list:
    
    sub_df = df[df[grp]>80].sort_values(by='30D Vol', ascending=False)
    
    if not sub_df.empty:
        etf_name = sub_df.index.values[0]
        grp_etfs.append((grp, etf_name))

grp_etfs = np.array(grp_etfs)
```
